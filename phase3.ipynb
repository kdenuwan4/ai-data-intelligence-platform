Matplotlib 

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python


# Load your normalized data
normalized_data = np.load("ai_data_normalized.npy")  

# Convert to DataFrame for easier handling
columns = ["Feature1", "Feature2", "Feature3"]
df = pd.DataFrame(normalized_data, columns=columns)
```

```python
#Line Plot
df.plot(marker='o', figsize=(8,5), title="All Features Trend")
plt.xlabel("Index")
plt.ylabel("Values")
plt.show()

```

```python
#Bar Plot

plt.bar(df.index, df["Feature1"])
plt.title("Feature1 Values")
plt.xlabel("Index")
plt.ylabel("Feature1")
plt.show()
```

```python
#Scatter Plot of Two Features

plt.scatter(df["Feature1"], df["Feature3"])
plt.title("Feature1 vs Feature3")
plt.xlabel("Feature1")
plt.ylabel("Feature3")
plt.show()
```

```python
#Histogram of a Feature
plt.hist(df["Feature3"], bins=5, edgecolor='black')
plt.title("Feature3 Distribution")
plt.show()
```
